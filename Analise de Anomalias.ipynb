# Outliers

## Método 1: IQR por estação de Metrô

```python
from scipy.stats import iqr

def outliers_iqr(serieEstacao):
    iqrDemanda = iqr(serieEstacao,nan_policy='omit')
    quartile_1, quartile_3 = np.percentile(serieEstacao.dropna().values, [25, 75])
    lower_bound = quartile_1 - (iqrDemanda * 1.5)
    upper_bound = quartile_3 + (iqrDemanda * 1.5)
    return np.where((serieEstacao > upper_bound) | (serieEstacao < lower_bound))

diaLinha1 = diaLinha.copy(deep=True)

todosDiasOutliers = []
for est in estMetro:
    outliersDestaEstacao = outliers_iqr(diaLinha[est])
    print(est,outliersDestaEstacao)
    todosDiasOutliers.append(outliersDestaEstacao)    
    diaLinha1.loc[:,est] = 0
    diaLinha1.loc[diaLinha1.iloc[outliersDestaEstacao].index,est] = 1
    
todosDiasOutliers = np.unique(np.hstack(todosDiasOutliers))
```

```python
diaLinha1[estacoes].to_csv('outliers metodo iqr.csv')
```

```python
diasNormais = np.setdiff1d(np.arange(0,len(diaLinha)),todosDiasOutliers)
diasNormais
```

## Método 2: Zscore por estação

```python
from scipy import stats

#Tira colunas com NaN (linhas que não comunicaram todo dia útil válido por algum motivo)
diaLinhaSemNan = diaLinha.loc[:,diaLinha.isnull().sum()==0]

#Verifica quais estações tem a série completa (tira inauguradas no período)
estacoesCompletas = np.intersect1d(estacoes,diaLinhaSemNan.columns)

diaLinhaEstacoesSemNan = diaLinhaSemNan[estacoesCompletas].copy(deep=True)

#Calcula o z index
z = np.abs(stats.zscore(diaLinhaEstacoesSemNan))

diasInvalidos = np.unique(np.where(z>3)[0])
```

```python
diaLinhaEstacoesSemNan1 = diaLinhaEstacoesSemNan.copy(deep=True)
diaLinhaEstacoesSemNan1[:] = 0

for i,j in zip(np.where(z>2)[0],np.where(z>2)[1]):
    diaLinhaEstacoesSemNan1.loc[diaLinhaEstacoesSemNan1.iloc[i].name,diaLinhaEstacoesSemNan1.columns.values[j]] = 1
```

```python
diaLinhaEstacoesSemNan1.to_csv('zscore2.csv')
```

```python
diasNormais = np.setdiff1d(np.arange(0,len(diaLinha)),diasInvalidos)
diasNormais
```

# Método 3: IsolationForest

```python
from sklearn.ensemble import IsolationForest
```

```python
diaLinhaEstacoesSemNan.values
```

```python
#Split train test
from sklearn.model_selection import train_test_split
rnd = np.random.RandomState(42)
estacoesValores = diaLinhaEstacoesSemNan.values
blt_train, blt_test = train_test_split(estacoesValores, test_size=0.33, random_state=rnd)

print(blt_train)
print(blt_test)
```

```python
#Aplica modelo IsolationForest
parametro=0.3
clf = IsolationForest(max_samples=100, random_state=rnd,contamination=parametro)
clf.fit(blt_train)
blt_pred_train = clf.predict(blt_train)
blt_pred_test = clf.predict(blt_test)
```
